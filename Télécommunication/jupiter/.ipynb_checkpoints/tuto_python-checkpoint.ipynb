<h3>utilisation des tableaux/liste/dictionnaire</h3>

- une liste/tableau:
    - contient uniquement des valeurs
    - l'ajout d'un nouvel élément dans la/le liste/tableau se fait à l'aide de la fonction **.append(\<élément\>)** 
    - la suppression d'une entrée dans un/e tableau/liste se fait à l'aide de la fonction **.pop(\<index\>)**  
    - fonctionne avec les **index négatifs**. -1 => récupère le dernier élément de la liste
- un dictionnaire:
    - fonctionne selon un système de **clef valeur** 
    - l'ajout d'information dans un dictionnaire fonctionne comme ceci: 
    ```python
    monDictionnaire["ma clef"] = "ma valeur"
    ```
    - la suppression d'une entrée clef valeur fonctionne aussi avec la fonction **.pop(\<nom de la clef\>)**



````python
print("----------------------TABLEAUX/LISTES----------------------\n")
tableau_ou_liste_de_chaine_de_caracteres.append("item3")
print("voici une liste contenant un item en plus {}".format(tableau_ou_liste_de_chaine_de_caracteres))
result=tableau_ou_liste_de_chaine_de_caracteres.pop(-1)
print("voici le résultat de la fonction pop '{}' et les informations contenues dans la liste {}".format(result,tableau_ou_liste_de_chaine_de_caracteres))
print("\n----------------------DICTIONNAIRES----------------------\n")
dictionnaire_de_characteres["clef"]=("valeur")
print("voici un dictionnairecontenant un item en plus {}".format(dictionnaire_de_characteres))
result=dictionnaire_de_characteres.pop("clef")
print("voici le résultat de la fonction pop '{}' et les informations contenues dans le dictionnaire {}".format(result,dictionnaire_de_characteres))
````

<h3>Branchement conditionnel</h3>

Cette section permet d'exécuter du code selon une ou plusieurs conditions plus ou moins complexe. 

Pour cela il faut utiliser la syntaxe **`if <condition>:`** où le mot clé **`if`** doit figurer en premier 


````python
valeurDynamique = int(input("veuillez rentrez un chiffre entier :"))
valeur = 56
 # Si la valeur valeurDynamique est plus petite que valeur alors
if valeurDynamique < valeur:
    print("la valeur rentrée par l'utilisateur ({}) est plus petite que la valeur du programme {}".format(valeurDynamique,valeur))
    print("moi aussi je veux apparaître sur la console...")
    print("OUI JE SUIS AFFICHER :)")
# Sinon si la valeur valeurDynamique est égale à 57
elif valeurDynamique == 57:
    print("la valeur rentrée par l'utilisateur ({}) == 57".format(valeurDynamique))
    print("Bonjour")
# sinon 
else:
    print(">={}".format(valeur))
    print("au revoir")
print("---------------------------END-----------------------")
````

Comme vous avez pu le constater, lorsqu'une condition `if` renvoie la valeur `True` on dit qu'elle est **vraie**, les commandes à éxécuter sont celles qui se trouve à **4 d'indentation** en rapport à la ligne du dessus (bloc de code). Dans le cas où la condition du `if` est **fausse** ou renvoie la valeur `False`, la condition `elif` va être exécutée et dans le cas où elle renvoie `True`, le bloc de code du `elif` s'éxécutera.

Dans les derniers cas, le `else` signifie qu'aucune condition n'est vraie et exécute dans ce cas le bloc de code correspondant.

**NB. un `if` n'est pas toujours suivi d'un `else` ou d'un `elif`**.

Pour verifier que vous ayez compris, essayez d'exécuter le code suivant. quelques erreurs d'indentation sont détectées. Corrigez les.

<h3>Multi-Conditions</h3>

Il est possible d'avoir plusieurs conditions sur un seul `if` en utilisant les mots clées `or` ou `and` pour combiné les conditions.
Une condition avec un `or` est vraie si l'une des deux conditions qui l'entoure est vraie
par ex.: 
```python
True or True # Renvoie True
True or False # Renvoie True
False or True # Renvoie True
False or False # Renvoie False
```

Une condition avec un `and` est vraie si *les deux conditions* qui l'entoure sont vraies
```python
True and True # Renvoie True
True and False # Renvoie False
False and True # Renvoie False
False and False # Renvoie False
```

Le mot clé `not` permet d'*inverser* la valeur de la condition par exemple:
```python
not True # Renvoie False
not (True and False) # Renvoie True
not (False or True) # Renvoie False
(not False and True) # Renvoie True
```

**NB. Les conditions se trouvant entre parenthèses sont évaluées avans les autres.***

````python
valeurDynamique = int(input("veuillez rentrez un chiffre entier :"))
if (valeurDynamique == 32 or valeurDynamique == 30) and valeurDynamique < 33:
    print("la valeur de l'utilisateur est soit 32 ou 30 et sa valeur est plus petite que 33")
    print("oui le if n'a aucun sens car la valeur doit être 32 ou 30 mais c'est obligatoire pour comprendre que\n 1. les parenthèses ont la priorité\n 2. c'est un TUTORIEL")
else:
    print("les conditions ne sont pas respectées")
````

### Boucles

Les boucles sont le nerf de la guerre en programmation. Elles permettent:

1. d'exécuter un block de commande *X* fois : (boucle `for`)
2. d'exécuter un block de commande selon le nombre d'élément dans un/e tableau/dicionnaire/liste (`foreach`)
3. d'exécuter un block de commande tant que la ou les conditions sont vrai : (`while`)

#### For/foreach

La différence entre une boucle for et une boucle foreach est que :

- la boucle `for` iter sur des nombres (tableau/liste)
- la boucle `foreach` iter sur des valeurs de tableau/dictionnaire

En python, il n'y pas de différence entre For et Foreach comme en JAVA!

````python
lst=["Luc", "je", "suis", "ton", "père"]
print("---------------------BOUCLE FOR SUR LISTE---------------------\n")

res=""
for i in range(0, len(lst)):
    res += "{} ".format(lst[i])

print(res)
print("\n---------------------BOUCLE FOREACH SUR LISTE---------------------\n")

res = ""   
for mot in lst:
    res +="{} ".format(mot)

print(res)

print("\n---------------------BOUCLE FOR SUR un Dictionnaire---------------------\n")
print("impossible car le dictionnaire fonctionne sur un système clef valeur")
print("\n---------------------BOUCLE FOREACH SUR UN Dictionnaire---------------------\n")

dic = {
    "je": "ceci",
    "suis": "est",
    "une": "un",
    "clef": "message sur les valeurs"
}

msgValeur = ""
i = 0
for value in dic.values():
    i = i + 1
    print(" test   ",(i))
    msgValeur += "{} ".format(value)
msgClef = ""

for clef in dic.keys():
    msgClef += "{} ".format(clef)

print("message valeurs: {} \nmessage clefs: {}".format(msgValeur, msgClef))

print("\nrécupérer clefs valeurs en même temps:\n")
for clef, valeur in dic.items():
    print("{} -> {}".format(clef, valeur))
````

#### Utiliser un `for` sur plusieur liste en même temps

il est bien sur possible de lire deux listes en même temps grâce à la fonction **zip(\<liste1\>,\<liste2\>)**


````python
lst1 = ["je", "lu", "même"]
lst2 = ["suis", "en", "temps"]
msg = ""
for v1, v2 in zip(lst1, lst2):
    msg += "{} {} ".format(v1, v2)
print(msg)    

for ind, v in enumerate(lst1):
    print("lst[{}]= {} ".format(ind, v))
````

<h3>Boucle while</h3>

la boucle while est exécutée tante que la condition est **vrai**. par exemple ma variable **i est égale à 0**. Tant que **i** est **plus petit que 10**, afficher **i**

````python
i = 0

while i < 10:
    print(i)
    i += 1
````
