```python
import math
import os

# exercice c
text_path = 'input.txt'
with open(text_path, 'r', encoding="UTF-8") as file:

    text = file.read()
    file.close()


def count_frequency(string, letter):
    counter = 0
    for character in string:
        if character == letter:
            counter += 1
    return counter


def CalculateEntropy(resultats_triee):
    H = 0
    for character in resultats_triee:
        H += character['apparition'] * -math.log2(character['apparition'])
    return H

# file1_name = input.txt file2_name = out2.txt


def CalculateCompression(file1_name, file2_name):
    # calcule avec la taille des fichiers textes
    file1_stats = os.stat(file1_name)
    file2_stats = os.stat(file2_name)

    file1_size = file1_stats.st_size
    file2_size = file2_stats.st_size
    compression_rate = file1_size/file2_size * 100

    # calcule avec la longueur du textes
    with open(file1_name, 'r', encoding="UTF-8") as file1:
        text1 = file1.read()
    with open(file2_name, 'r', encoding="UTF-8") as file2:
        text2 = file2.read()

    compression_rate = len(text2) / len(text1) * 100
    file1.close()
    file2.close()
    return compression_rate


# Extraire les lettres du texte
letters = [char for char in text]
# Trier les lettres en utilisant leur valeur ASCII
letters.sort(key=lambda x: ord(x))

resultats = []

letter_uniques = list(set(letters))

for letter in letter_uniques:
    letter_Info = {  # un crée un dictionnaire
        'character': letter,
        'hexa': hex(ord(letter)),
        'apparition': round(count_frequency(letters, letter)/len(text), 3)
    }
    resultats.append(letter_Info)

resultats_triee = sorted(
    resultats, key=lambda x: x['apparition'], reverse=True)


text_result = ""

for key in resultats_triee:
    text_result += str(key['character']) + " " + \
        key['hexa'] + " " + str(key['apparition']) + " \n"

D = math.log2(len(text))
H = CalculateEntropy(resultats_triee)
Redondance = D-H
compression = CalculateCompression('input.txt', 'out2.txt')

# création d'une class noeud avec comme proprités une valeur (ASCII) + sa frequence + noeuds enfants(gauche et droites)
class Node:
    def __init__(self, value, freq):
        self.value = value
        self.freq = freq
        self.left = None
        self.right = None

# fonction pour construire l'arbre de huffman

def build_huffman_tree(resultats_triee):
    #initialisation des noeuds
    nodes = [Node(ord(characterInfo['character']),characterInfo['apparition']) for characterInfo in resultats_triee]
    while len(nodes) > 1:
        # Triez la liste de nœuds par fréquence (ordre croissant)
        nodes.sort(key=lambda x: x.freq)
        
        # Prenez les deux nœuds avec les fréquences les plus basses
        left = nodes.pop(0)
        right = nodes.pop(0)
        
        # Créez un nouveau nœud parent avec la somme des fréquences
        parent = Node(None, left.freq + right.freq)
        parent.left = left
        parent.right = right
        
        # Ajoutez le nœud parent à la liste
        nodes.append(parent)
    # Le dernier nœud restant est la racine de l'arbre
    return nodes[0]

#root = le noeud racine de l'arbre de Huffman, code = string qui représente le code actuel en cours de construction, mapping est un dictionnaire ou nous stockerons les codes 
# de huffman à chaque valeur
def generate_huffman_codes(root, code, mapping):
        # Vérifiez si le nœud actuel (root) est None (fin de l'arbre).
    if root:
        # Si le nœud actuel a une valeur (c'est une feuille de l'arbre).
        if root.value:
            # Associez cette valeur (feuille) avec le code actuel dans le dictionnaire mapping.
            mapping[root.value] = code
        # Appel récursif pour le sous-arbre gauche avec un code modifié (ajout de "0").
        generate_huffman_codes(root.left, code + "0", mapping)
        # Appel récursif pour le sous-arbre droit avec un code modifié (ajout de "1").
        generate_huffman_codes(root.right, code + "1", mapping)


huffman_tree = build_huffman_tree(resultats_triee)
huffman_codes = {}
generate_huffman_codes(huffman_tree, '', huffman_codes)

for value, code in huffman_codes.items():
    print(f"{hex(value)}: {code}")
```
